### SETTING 200 gain

```python
%matplotlib inline
dset = WBDataset(dsname='Fresnel_Suzukii_RL/TEST200gain16Khz', clean=False, custom_label=[0])

r_dset = random.choices(dset, k=25)

plt.figure(figsize=(26,12))
for i in range(len(r_dset)):
    plt.subplot(5,5,i+1)
    try:
        sig = read_wbt(r_dset[i]['path'], filtered=True)
        sig = sig[:10000] # 0.625 seconds (sampled at 16khz)
        sig =  signal.resample(sig, 5000)

        plt.plot(sig); plt.ylim(-.15,.15); plt.title(len(sig))
    except:
        continue
```
